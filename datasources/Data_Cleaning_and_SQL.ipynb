```python
import pandas as pd
import math
from sqlalchemy import create_engine
from config import password
import os
```

```python
# import csv file
data = pd.read_csv("Sex_Offender_Registry.csv")
```

```python
# display full data in table
pd.set_option('display.max_columns', None)
```

```python
data
```

```python
# change datetime format to only include date
clean_data = data

for i in range(len(clean_data)):
    clean_data.iloc[i, 5] = clean_data.iloc[i, 5][:10]
    clean_data.iloc[i, 7] = clean_data.iloc[i, 7][:10]
    clean_data.iloc[i, 12] = clean_data.iloc[i, 12][:10]
    clean_data.iloc[i, 18] = clean_data.iloc[i, 18][:10]
    if math.isnan(clean_data.iloc[i, 20]):
        pass
    else:
        clean_data.iloc[i, 20] = int(clean_data.iloc[i, 20])
```

```python
clean_data = clean_data.rename(columns={
    'X': 'Longitude',
    'Y': 'Latitude',
    'OBJECTID': 'Object_ID',
    'BLOCK_X': 'Block_X',
    'BLOCK_Y': 'Block_Y',
    'DCS_LAST_MOD_DTM': 'Last_Updated',
    'SEXOFFENDERCODE': 'Sex_Offender_Code',
    'ADDDATE': 'Date_Added',
    'LASTNAME': 'Last_Name',
    'FIRSTNAME': 'First_Name',
    'ALIASES': 'Aliases',
    'MAXCLASSIFICATION': 'Max_Classification',
    'BIRTHDATE': 'DOB',
    'HEIGHTNUM': 'Height_Inches',
    'WEIGHTNUM': 'Weight_lbs',
    'EYECOLOR': 'Eye_Color',
    'HAIRCOLOR': 'Hair_Color',
    'MARKINGS': 'Markings',
    'REGISTRATIONDATE': 'Date_Registered',
    'TYPE': 'Residence',
    'DISTRICT': 'District',
    'PSA': 'PSA_Announcement_Number',
    'BLOCKNAME': 'Block_Address',
    'QUADRANT': 'Quadrant',
    'ZIP': 'ZIP_Code'
})
```

```python
clean_data
```

```python
# Clean dataset to only include unique offenders
no_duplicate_data = clean_data.drop_duplicates(subset=['Sex_Offender_Code'])
```

```python
# export dataset with unique add dates for line chart
no_duplicate_data.to_csv("unique_add_dates.csv")
```

## Export data into SQL

```python
rds_connection_string = f"postgres:{password}@localhost:5432/sex_offender_db"
engine = create_engine(f"postgresql://{rds_connection_string}")
```

```python
engine.table_names()
```

```python
clean_data.to_sql(name='registry', con=engine, if_exists='append', index=False)
no_duplicate_data.to_sql(name='unique_dates', con=engine, if_exists='append', index=False)
```
